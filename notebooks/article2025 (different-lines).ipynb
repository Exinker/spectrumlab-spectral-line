```python
import json

from spectrumlab.lines import Line

from spectrumlab_spectral_line.constants import SIGMA, TAU
from spectrumlab_spectral_line.estimators import LineShapeEstimator, LineShapeEstimatorConfig


lines = [
    ('Al', 308.2153),
    ('Al', 309.27099),
    ('Al', 394.40058),
    ('Cs', 455.5275),
    ('Cs', 852.1),
    ('Mg', 202.5824),
    ('Mg', 285.2127),
    ('Ca', 227.5466),
    ('Ca', 422.6728),
    ('Zn', 213.8573),
    ('Zn', 307.5895),
    ('Na', 330.2978),
    ('Na', 588.995),
    ('K', 404.4142),
    ('K', 766.49),
]
```

```python
estimator = LineShapeEstimator(
    config=LineShapeEstimatorConfig(
        temperature=2400 + 273.15,
        buffer='Ar',
        tau=TAU,
        sigma=SIGMA,
    ),
)

shapes = []
for element, wavelength in lines:
    shape = estimator.fit(line=Line(element, wavelength))
    shapes.append(dict(
        line=f'{element} {wavelength}',
        width=f'{shape.width:.4f}',
        ratio=f'{shape.ratio:.4f}',
        hwhm=f'{shape.fwhm:.4f}',
    ))
```

```python
print(json.dumps(shapes, indent=2))
```
